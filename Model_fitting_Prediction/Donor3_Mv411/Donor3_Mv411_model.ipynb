```python
LB = np.array([0.07,0.01,500,1.0e-1,1.0e-1,
               1.0e-2,
               1.0e-2,1.0e-2,
               1.e-6, 1.e-4])
UB = np.array([0.1,0.04,4700,1.0,1,
               1.0e+0,
               1.0e+0,1.0e+0,
               1.e-4,1.0e-2])
```

```python
def main_Mv411():
    Tr = 46500.0
    Sys_CAR_NK.Cell_type_R_H(Targets = Tr)
    Sys_WT_NK.Cell_type_R_H(Targets = Tr)
    t = 48.0
    x0 = np.array([0.0713213354195174, 0.039914365733240253, 1536.0902176019104, 0.10000506703270665, 0.4423226396919762, 0.6723505447438678, 0.16875886803032464, 0.010000002132358672, 7.394127647073689e-06, 0.003915253798958188])
    res0 = Model_fit(x0,LB,UB,Sys_CAR_NK,Sys_WT_NK,WT_NK,v5_NK,v6_NK,t,fit=False)
    return res0
```
