```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as  plt
import matplotlib.colors as mcolors
```

### ET_Ratio 0.1

```python
df0 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_a4_ET_1.xlsx')
df0.sort_values(by='w', inplace=True)
df1 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_a4_ET_5.xlsx')
df1.sort_values(by='w', inplace=True)
df2 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_a4_ET_10.xlsx')
df2.sort_values(by='w', inplace=True)
```

```python
cmap = 'coolwarm'
norm = mcolors.Normalize(vmin=0.0, vmax=1.0)
fig,ax = plt.subplots(1,1,figsize = (8,6))
ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=18, length=6, width=5)
plt.rcParams['axes.linewidth'] = 5
markers = ['o','d','^','*']
size = 30
x = np.linspace(0,80,100)
y = np.linspace(0,80,100)
ax.plot(x,y,color = 'gray',lw=2.5,alpha=0.85)
sc = ax.scatter(df0['Healthy_Lysis'], df0['Tumor_lysis'], marker='o', s = size, c=df1['w'], cmap=cmap,norm=norm,label = 'ET 1:1')
sc = ax.scatter(df1['Healthy_Lysis'], df1['Tumor_lysis'], marker='d', s = size, c=df1['w'], cmap=cmap, norm=norm,label = 'ET 5:1')
sc = ax.scatter(df2['Healthy_Lysis'], df2['Tumor_lysis'], marker='^', s = size, c=df2['w'], cmap=cmap, norm=norm,label = 'ET 10:1')
cbar = fig.colorbar(sc)
#cbar.set_ticks()  # Set the positions of the colorbar ticks
#cbar.set_ticklabels(ET_ratio)
#plt.xlabel('%Lysis Healthy cells',fontsize=18)
#plt.ylabel('%Lysis Tumor cells',fontsize=18)
plt.xticks([0.0,40.0,80.0],fontsize=20)
plt.yticks([0.0,40.0,80.0],fontsize=20)
plt.legend(fontsize=10.8,loc='lower right', bbox_to_anchor=(1.0, 0.0))
#plt.title("Pareto Front with Model Min_x Max_y_fixed_a1_a4")
plt.savefig('Plot_Pareto_with_Model_Min_x_Max_y_fixed_a1_a4.png',bbox_inches='tight')
plt.show()
```

```python
df0 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_a1_0.25_ET_1.xlsx')
df0.sort_values(by='w', inplace=True)
df1 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_a1_0.25_ET_5.xlsx')
df1.sort_values(by='w', inplace=True)
df2 = pd.read_excel('a4_bw_p001_1_Lysis_Kasumi_Healthy_CAR_NK_a1_0.25_ET_10.xlsx')
df2.sort_values(by='w', inplace=True)

cmap = 'coolwarm'
norm = mcolors.Normalize(vmin=0.0, vmax=1.0)
fig,ax = plt.subplots(1,1,figsize = (8,6))
ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=18, length=6, width=1.5)
plt.rcParams['axes.linewidth'] = 3
markers = ['o','d','^','*']
size = 50
x = np.linspace(0,80,100)
y = np.linspace(0,80,100)
ax.plot(x,y,color = 'gray')
sc = ax.scatter(df0['Healthy_Lysis'], df0['Tumor_lysis'], marker='o', s = size*df0['alpha4'], c=df0['w'],cmap=cmap,label = 'ET 1:1')
#sc = ax.scatter(df1['Healthy_Lysis'], df1['Tumor_lysis'], marker='d', s = size*df1['alpha4'], c=df1['w'], cmap=cmap, norm=norm,label = 'ET 1:5')
#sc = ax.scatter(df2['Healthy_Lysis'], df2['Tumor_lysis'], marker='^', s = size*df2['alpha4'], c=df2['w'], cmap=cmap, norm=norm,label = 'ET 1:10')
cbar = fig.colorbar(sc, label='w')
#cbar.set_ticks()  # Set the positions of the colorbar ticks
#cbar.set_ticklabels(ET_ratio)
plt.xlabel('%Lysis Healthy cells',fontsize=16)
plt.ylabel('%Lysis Tumor cells',fontsize=16)
plt.xticks([0.0,40.0,80.0])
plt.yticks([0.0,40.0,80.0])
plt.legend(fontsize=10.8,loc='lower right', bbox_to_anchor=(1.0, 0.0))
plt.title("Pareto Front with Model Min_x Max_y_a1_0.25")
plt.savefig('Plot_Pareto_with_Model_Min_x_Max_y_ET1.png',bbox_inches='tight')
plt.show()
```
